**Cách hoạt động của giải thuật:**

> Bubble sort (Sắp xếp nổi bọt) duyệt qua danh sách và so sánh các cặp phần tử liền kề. Các phần tử được hoán đổi nếu chúng không đúng thứ tự. Việc chuyển qua phần chưa được sắp xếp của danh sách được lặp lại cho đến khi danh sách được sắp xếp.

> 


```python
def bubble_sort(nums):
    for j in range(len(nums)-1,0,-1):
        for i in range(j):
            if nums[i] > nums[i+1]:
                nums[i], nums[i+1] = nums[i+1], nums[i]
                
list_num = [54,26,93,17,77,31,44,55,20]
bubble_sort(list_num)
print(list_num)
```
