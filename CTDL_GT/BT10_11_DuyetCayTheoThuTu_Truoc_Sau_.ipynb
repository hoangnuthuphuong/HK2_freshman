**Cách hoạt động của giải thuật duyệt cây theo thứ tự trước**

 Duyệt cho tới khi tất cả các nút đều được duyệt:

> **Duyệt trung thứ tự**

Bước 1: Duyệt các cây con bên trái một cách đệ qui

Bước 2: Truy cập nút gốc

Bước 3: Duyệt các cây con bên phải một cách đệ qui

> **Duyệt theo thứ tự trước**

Bước 1: Truy cập nút gốc

Bước 2: Duyệt các cây con bên trái một cách đệ qui

Bước 3: Duyệt các cây con bên phải một cách đệ qui

> **Duyệt theo thứ tự sau**

Bước 1: Duyệt các cây con bên trái một cách đệ qui

Bước 2: Duyệt các cây con bên phải một cách đệ qui

Bước 3: Truy cập nút gốc

```python
class Node:
    """
    Class Node
    """
    def __init__(self, value):
        self.left = None
        self.data = value
        self.right = None
 
class Tree:

   # Hàm để tạo node
    def createNode(self, data):
        return Node(data)
   # Hàm để chèn các node vào cây
    def insert(self, node , data):
        
        # Nếu cây trống -> trả về nút gốc
        if node is None:
            return self.createNode(data)
        # Nếu dữ liệu nhỏ hơn dữ liệu gốc -> chèn nó vào bên trái
        if data < node.data:
            node.left = self.insert(node.left, data)
        # Ngược lại -> chèn vào bên phải
        elif data > node.data:
            node.right = self.insert(node.right, data)
        return node
 
    # Hàm tìm kiếm để tìm kiếm node thành cây
    def search(self, node, data):
        if node is None or node.data == data:
            return node
        if node.data < data:
            return self.search(node.right, data)
        else:
            return self.search(node.left, data)
 
    # Hàm ìn tất cả các node trong cây
    # Duyệt trung thứ tự In-order Traversal 
    def traverseInorder(self, root):
        if root is not None:
            self.traverseInorder(root.left)
            print(root.data)
            self.traverseInorder(root.right)
    # Duyệt thứ tự trước Pre-order Traversal 
    def traversePreorder(self, root):
        if root is not None:
            print(root.data)
            self.traversePreorder(root.left)
            self.traversePreorder(root.right)
    # Duyệt thứ tự sau Post-order Traversal 
    def traversePostorder(self, root):
        if root is not None:
            self.traversePreorder(root.left)
            self.traversePreorder(root.right)
            print(root.data)
 
 
def main():
    root = None
    tree = Tree()
    root = tree.insert(root, 5)
    print(root)
    tree.insert(root, 100)
    tree.insert(root, 17)
    tree.insert(root, 9)
    tree.insert(root, 15)
    tree.insert(root, 5)
    tree.insert(root, 11)
 
    print("Duyệt cây trung thứ tự - Inorder")
    tree.traverseInorder(root)
 
    print("Duyệt cây theo thứ tự sau - Preorder")
    tree.traversePreorder(root)
 
    print("Duyệt cây theo thứ tự sau - Postorder")
    tree.traversePostorder(root)
 
 
if __name__ == "__main__":
    main()
```
