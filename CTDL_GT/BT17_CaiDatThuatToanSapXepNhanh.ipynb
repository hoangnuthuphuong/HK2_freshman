**Cách hoạt động của giải thuật:**

> Sắp xếp nhanh (Quicksort), còn được gọi là sắp xếp trao đổi phân vùng (phân vùng trao đổi), chia dữ liệu được sắp xếp thành hai phần độc lập thông qua một cách sắp xếp, tất cả dữ liệu ở một phần nhỏ hơn tất cả dữ liệu ở phần kia Và sau đó theo phương pháp này để nhanh chóng sắp xếp hai phần của dữ liệu, toàn bộ quá trình sắp xếp có thể đệ quy, để đạt được toàn bộ dữ liệu thành một chuỗi có thứ tự.

> Các bước:

* Chọn một phần tử từ trình tự và gọi nó là "Pivot"

* Sắp xếp lại trình tự, tất cả các phần tử nhỏ hơn giá trị điểm chuẩn được đặt ở phía trước điểm chuẩn và tất cả các phần tử lớn hơn giá trị điểm chuẩn được đặt phía sau điểm chuẩn (cùng một số có thể nằm ở hai bên). Sau khi phân vùng kết thúc, điểm chuẩn nằm ở giữa chuỗi. Đây được gọi là hoạt động phân vùng.

* Sắp xếp đệ quy các dãy con của phần tử nhỏ hơn giá trị tham chiếu và dãy con của các phần tử lớn hơn giá trị tham chiếu.

```python
def quick_sort(alist, start, end):

    if start >= end:
        return

    mid = alist[start]

    low = start

    high = end

    while low < high:
        while low < high and alist[high] >= mid:
            high -= 1
        alist[low] = alist[high]

        while low < high and alist[low] < mid:
            low += 1
        alist[high] = alist[low]

    alist[low] = mid

    quick_sort(alist, start, low-1)

    quick_sort(alist, low+1, end)


alist = [3,5,33,1,8,12,4,23,8]
quick_sort(alist,0,len(alist)-1)
print(alist)
```
