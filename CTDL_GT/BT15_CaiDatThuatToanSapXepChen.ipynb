**Cách hoạt động của giải thuật:**

> Insertion sort (Sắp xếp chèn) giống như cách nhiều người sắp xếp các quân bài của họ khi chơi bài tiến lên.

> Thực hiện sắp xếp dãy số theo cách duyệt từng phần tử và chèn từng phần tử đó vào đúng vị trí trong mảng con(dãy số từ đầu đến phần tử phía trước nó) đã sắp xếp sao cho dãy số trong mảng sắp đã xếp đó vẫn đảm bảo tính chất của một dãy số tăng dần.
